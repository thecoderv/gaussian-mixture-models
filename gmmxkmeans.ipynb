```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
# mean and cov for dataset1
mean1 = [-1.0, -1.5]
cov1 = [[1.0, 0.2], [0.2, 1]]
```

```python
# generating dataset1
x1, y1 = np.random.default_rng().multivariate_normal(mean1, cov1, 100).T
plt.plot(x1, y1, 'x')
plt.axis('equal')
plt.title("Class 1")
plt.show()
np.asarray(x1)
np.asarray(y1)
data1 = np.zeros((100,2))
data1[:,0] = x1
data1[:,1] = y1
#print(data1)
```

```python
# mean and cov for dataset2
mean2 = [1.0, 1.5]
cov2 = [[2.0, 0.1], [0.1, 2.0]]
```

```python
# generating dataset2
x2, y2 = np.random.default_rng().multivariate_normal(mean2, cov2, 100).T
plt.plot(x2, y2, 'x')
plt.axis('equal')
plt.title("Class 2")
plt.show()
np.asarray(x2)
np.asarray(y2)
data2 = np.zeros((100,2))
data2[:,0] = x2
data2[:,1] = y2
#print(data2)
```

```python
Class1 = np.ones((100,1))
Class2 = np.full((100,1), 2)
Cdata1 = np.c_[data1, Class1]
df1 = pd.DataFrame(Cdata1, columns=["X1", "X2", "labels"])
Cdata2 = np.c_[data2, Class2]
df2 = pd.DataFrame(Cdata2, columns=["X1", "X2", "labels"])
dfT = df1.append(df2, ignore_index=True)
```

```python
plt.figure(0)
sns.pairplot(dfT, hue="labels", palette="Set2")
plt.show()
```

```python
dataf = np.zeros((200,2))
dataf = np.concatenate((data1, data2), axis=0)
```

```python
#GMM clustering for dataset
gmm1 = GaussianMixture(n_components=4)
gmm1.fit(dataf)
labels1 = gmm1.predict(dataf)
frame1 = pd.DataFrame(dataf)
frame1['cluster'] = labels1
frame1.columns = ['x1', 'y1', 'cluster']
#print(frame1)
```

```python
color=['blue','green','cyan', 'black', 'red', 'yellow']
for k in range(0,4):
    d1 = frame1[frame1["cluster"]==k]
    plt.scatter(d1["x1"],d1["y1"],c=color[k])
plt.title("GMM Clustering")
plt.show()
```

```python
#Kmeans for dataset
kmeans1 = KMeans(n_clusters=4)
kmeans1.fit(dataf)
pred1 = kmeans1.predict(dataf)
frame3 = pd.DataFrame(dataf)
frame3['cluster'] = pred1
frame3.columns = ['x1', 'y1', 'cluster']
#print(frame3)
```

```python
color=['blue','green','cyan', 'black','red', 'yellow']
for k in range(0,4):
    d1 = frame3[frame3["cluster"]==k]
    plt.scatter(d1["x1"],d1["y1"],c=color[k])
plt.title("KMeans Clustering")
plt.show()
```

```python
sns.pairplot(frame1, hue="cluster", palette="Set2")
plt.suptitle('GMM clustering')
sns.pairplot(frame3,hue="cluster", palette="Set2")
plt.suptitle('KMeans clustering')
```
